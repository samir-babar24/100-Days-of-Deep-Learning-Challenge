```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

# Normal Initialization

```python
model1=Sequential()

model1.add(Dense(10,input_dim=2,activation="tanh",kernel_initializer="glorot_normal"))
model1.add(Dense(10,activation="tanh",kernel_initializer="glorot_normal"))
model1.add(Dense(10,activation="tanh",kernel_initializer="glorot_normal"))
model1.add(Dense(10,activation="tanh",kernel_initializer="glorot_normal"))
model1.add(Dense(1,activation="sigmoid"))

model1.summary()
```

```python
print(model1)
```

```python
model1.get_weights()
```

```python
model1.get_weights()
```

```python
model1.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model1.fit(x,y,epochs=30,validation_split=0.2)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x.values,y.astype('int').to_numpy(), clf=model1, legend=2)
```
