```python
import time
def fib(n):
   if n==1 or n==0:
     return 1

   else :
    return fib(n-1) + fib(n-2)
```

```python
#without memoization
start=time.time()
fib(38)
print(time.time()-start)
```

```python
#code with memoization:
import time
def fib(n,d):
   if n in d:
     return d[n]

   else :
    d[n]=fib(n-1,d) + fib(n-2,d)
    return d[n]

start1= time.time()
d={0:1,1:1}
fib(38,d)
print(time.time()-start1)
```

```python
#here we have saved a lot of time
```
