```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import layers ,Sequential
from tensorflow.keras.layers import Dense ,Flatten
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt
```

```python
(xtr,ytr),(xte,yte)=keras.datasets.mnist.load_data()
```

```python
'''here we have maximumm value as 253 in array so to make them in 0 to 1 we must need to divide them
253 , so lets divide them'''
```

```python
xtr=xtr/253
xte=xte/253
```

```python
xtr[0]
```

```python
#building the model
model=Sequential()#initialiing the model
model.add(Flatten(input_shape=(28,28))) #converting 2d ro 1d
model.add(Dense(128,activation='relu')) #hidden layer
model.add(Dense(10,activation='softmax')) #output and using doftmaxt for multiclass
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
history=model.fit(xtr,ytr,epochs=10,validation_split=0.2)
```

```python
y_prob=model.predict(xte)
```

```python
y_prob.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(yte,y_prob.argmax(axis=1))
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.legend(['loss','val_loss'])
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.legend(['Accuracy','val_Accuracy'])
plt.show()
```
